```python
def page_ready(driver):
    time.sleep(5)
    num = []
    for i in re.findall('[0-9]+', driver.find_element_by_id('navega').text.split(',')[0]):
        num.append(i)
    
    while not any(re.findall('[0-9]+', driver.find_element_by_id('navega').text)[0]):
        time.sleep(2)
    else:
        print('')
        
    entries_found = len(driver.find_elements_by_class_name('NoI'))
    while entries_found < int(max(num)):
        time.sleep(1)
        entries_found = len(driver.find_elements_by_class_name('NoI'))
    else:
        print("page ready")
```

```python
def nextpage_read(driver):
    time.sleep(5)
    num = []
    for i in re.findall('[0-9]+', driver.find_element_by_id('navega').text.split(',')[0]):
        num.append(i)
    num_page1 = max(num)
    num_page2 = int(re.findall('[0-9]+', driver.find_element_by_id('navega').text.split(',')[1])[0]) - int(num_page1)
    
    entries_found = len(driver.find_elements_by_class_name('NoI'))
    while num_page2 < num_page2:
        time.sleep(1)
        entries_found = len(driver.find_elements_by_class_name('NoI'))
    else:
        print("page ready")
```

```python
def obtain_metadata(driver, party):
    all_data = []
    
    years = []
    tipo_sesion = []
    periodo = []
    fecha = []
    diario = []
    tema = []
    link_speech = []
    
    for i in driver.find_element_by_id('output').find_elements_by_class_name('YearD'):
        years.append(i.text)
    
    #for i in driver.find_element_by_id('output').find_elements_by_class_name('TipoL'):
     #   tipo_sesion.append(i.text)
    
    #for i in driver.find_element_by_id('output').find_elements_by_class_name('PerL'):
     #   periodo.append(i.text)
        
    for i in driver.find_element_by_id('output').find_elements_by_class_name('FechaD'):
        fecha.append(i.text)
        
    #for i in driver.find_element_by_id('output').find_elements_by_class_name('DDD'):
     #   diario.append(i.text)
        
    for i in driver.find_element_by_id('output').find_elements_by_class_name('TemaI'):
        tema.append(i.text)
        link_speech.append(i.find_element_by_css_selector('a').get_attribute('href'))
    df = pd.DataFrame({
        'year_session':years, 
         #               'tipo_sesion':tipo_sesion, 
          #              'periodo':periodo, 
                        'fecha': fecha, 
          #              'diario': diario, 
                         'tema':tema,
                         'link_speech':link_speech
                        })
    df['partido'] = party
    df['name'] = driver.find_element_by_id('Escudo').text
    return(df)
```

```python
def obtain_allData(driver, partido):
    names_deputies_party = []
    partido_list = []
    for i in driver.find_element_by_id('selectDipu').find_elements_by_css_selector('option'):
        names_deputies_party.append(i.text)
        partido_list.append(partido)
        
    all_d = []
    for i, e in tqdm(zip(driver.find_element_by_id('selectDipu').find_elements_by_css_selector('option'), names_deputies_party), total = len(names_deputies_party)):
        i.click()
        time.sleep(5)
        all_d.append(load_all_pages(driver, partido))
    return(all_d)
```
