```python
import json
import pandas as pd
from os import listdir
```

```python
def read_json(path, list_):
    all_data = []
    for i in list_:
        if i == '.DS_Store' or 
            continue
        else:
            with open(path + i) as f:
                temp = json.load(f)
                for j in temp:
                    all_data.append(j)
    return(all_data)
            
    
```
