```python
import pandas as pd
import numpy as np
import os
import json
import unidecode
import re
from os import listdir
```

```python
path = '../../../data/01-collection_data/03-legislation/02-id_members/'
```

```python
all_data = []
for i in listdir(path):
    file = path + i
    try:
        with open(file) as f:
            data = json.load(f)
            for j in data:
                all_data.append(j)
    except:
        print(i)
        
```

```python
types_status = ['prorroga','publicado', 'desechada', 'retirada', 'turnada', 'turnado',
               'dictaminada en sentido positivo', 'dictaminada en sentido negativo', 
               'devuelta', 'turno modificado', 'concluido', 'considerada', 
               'pasa al ejecutivo federal', 'dictamen']
```

```python
def clean_type(x):
    clean_status = unidecode.unidecode(x.lower())
    types_status = ['prorroga','publicado', 'desechada', 'retirada', 'turnada', 'turnado',
               'dictaminada en sentido positivo', 'dictaminada en sentido negativo', 
               'devuelta', 'turno modificado', 'concluido', 'considerada', 
               'pasa al ejecutivo federal', 'dictamen']
    
    return(re.findall('|'.join(types_status), clean_status)[0])
```

```python
for i in pd.DataFrame(all_data).status_raw.unique():
    print(clean_type(i))
```
