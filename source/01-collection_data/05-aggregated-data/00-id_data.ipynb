```python
def clean_parties(x):
    if x == 'pan' or x == 'PAN':
        return('Partido Accion Nacional')
    elif x == 'prd' or x == 'PRD':
        return('Partido de la Revolucion Democratica')
    elif x == 'pri' or x == 'PRI':
        return('Partido Revolucionario Institucional')
    elif x == 'mc' or x == 'MC' or x == 'movimiento ciudadano':
        return('Movimiento Ciudadano')
    elif x == 'pve' or x == 'partido verde ecologista' or x == 'PVEM':
        return('Partido Verde Ecologista de Mexico')
    elif x == 'morena' or x == 'MORENA' or x =='morenal':
        return('Morena')
    elif x == 'encuentro social' or x == 'ES':
        return('Partido Encuentro Social')
    elif x == 'pt' or x == 'PT':
        return('Partido del Trabajo')
    elif x == 'NA' or x == 'sin partido':
        return('Sin Partido')
    elif x == 'alianza':
        return('Nueva Alianza')
    elif x == 'psd':
        return('Partido Social Democrata')
    else:
        return(x)
```

```python
import pandas as pd
import numpy as np
import os
import json
from os import listdir
```

```python
path = '../../../data/01-collection_data/00-id_data/00-id/'
```

```python
all_data = []
for i in listdir(path):
    file = path + i
    try:
        with open(file) as f:
            data = json.load(f)
            for j in data:
                j['partido'] = clean_parties(j['partido'])
                all_data.append(j)
    except:
        print(i)
        
```
