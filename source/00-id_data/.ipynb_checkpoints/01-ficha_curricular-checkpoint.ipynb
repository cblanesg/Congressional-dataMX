```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver import Chrome
from selenium import webdriver
import re
import time
import pandas as pd
import datetime
import uuid
from tqdm import tqdm
from datetime import date
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument('--headless')
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('http://sitl.diputados.gob.mx/LXIV_leg/curricula.php?dipt=57')
```

```python
table = driver.find_element_by_css_selector('body > div > table:nth-child(7) > tbody > tr > td > table:nth-child(2) > tbody > tr > td > div.curricula > table > tbody')
```

```python
temp = {}
for i in table.find_elements_by_css_selector('tr'): ## loop in rows
    if len(i.find_elements_by_css_selector('td')) == 1:
        titulo = i.find_elements_by_css_selector('td')[0].text
        temp[titulo] = []
        print(titulo)
    else:
        data_exp = i.find_elements_by_css_selector('td')[0].text + '|' + i.find_elements_by_css_selector('td')[1].text + '|' + i.find_elements_by_css_selector('td')[2].text 
        temp[titulo].append(data_exp)
        
        
df = []
for x, y in temp.items():
    for i in y:
        puesto, area, fecha = i.split('|')
        temp2 = {}
        temp2['legislatura'] = 64
        temp2['name_inc'] = 'camila'
        temp2['area_experiencia'] = x
        temp2['grado'] = puesto
        temp2['institucion'] = area
        temp2['tiempo'] = fecha
        df.append(temp2)
```

```python
def get_experience(driver, legis):
    temp = {}
    for i in table.find_elements_by_css_selector('tr'): ## loop in rows
        if len(i.find_elements_by_css_selector('td')) == 1:
            titulo = i.find_elements_by_css_selector('td')[0].text
            temp[titulo] = []
        else:
            data_exp = i.find_elements_by_css_selector('td')[0].text + '|' + i.find_elements_by_css_selector('td')[1].text + '|' + i.find_elements_by_css_selector('td')[2].text 
            temp[titulo].append(data_exp)


    df = []
    for x, y in temp.items():
        for i in y:
            puesto, area, fecha = i.split('|')
            temp2 = {}
            temp2['legislatura'] = legis
            temp2['name_inc'] = re.sub('Dip. ', '', driver.find_element_by_css_selector('body > div > table.cajasombra > tbody > tr > td > table > tbody > tr > td:nth-child(3) > table > tbody > tr:nth-child(1) > td > center > strong').text)
            temp2['area_experiencia'] = x
            temp2['grado'] = puesto
            temp2['institucion'] = area
            temp2['tiempo'] = fecha
            df.append(temp2)
    return(df)
```

```python
get_experience(driver, 'legis')
```
